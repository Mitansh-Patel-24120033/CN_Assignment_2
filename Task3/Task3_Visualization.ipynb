```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import json
```

```python
#Visualizing Task 3 Metrics
fl='metrics_3.json'
with open(fl, "r") as f:
    json_data = json.load(f)
data = pd.DataFrame(json_data)

data['configuration'] = data.apply(lambda row: f"Nagle: {row['nagle']}, Delayed-ACK: {row['delayed_ack']}", axis=1)

x = np.arange(len(data))  # Numeric positions for configurations

bar_width = 0.3
plt.figure(figsize=(10, 6))

# Throughput bars
plt.bar(x - bar_width / 2, data['throughput'], bar_width, label='Throughput', color='blue')
# Goodput bars
plt.bar(x + bar_width / 2, data['goodput'], bar_width, label='Goodput', color='yellow')
# Max Packet Size as a line plot
plt.plot(x, data['max_pac'], label='Max Packet Size', color='red', marker='o')

# Add title and labels
plt.title('TCP Performance Metrics by Configuration')
plt.xlabel('Configuration')
plt.ylabel('Metrics (Bytes)')
plt.xticks(x, data['configuration'], rotation=45, ha='right')
plt.legend()
plt.show()
```
