```python
import json
import matplotlib.pyplot as plt
import pandas as pd

def load_json(filename):
    with open(filename, 'r') as f:
        data = json.load(f)
        return pd.DataFrame(data)

# Load data with validation
nf_df = load_json('metrics_2_nf.json')
f_df = load_json('metrics_2_f.json')

# Sort data by client ID for proper plotting
nf_df = nf_df.sort_values('id')
f_df = f_df.sort_values('id')

# Create comparison plots
plt.figure(figsize=(10, 6))

plt.plot(nf_df['id'], nf_df['num_msgs'], 'b-o', label='Non-Flooded', markersize=3.5)
plt.plot(f_df['id'], f_df['num_msgs'], 'r-o', label='Flooded', markersize=3.5)
plt.title('Message Count Comparison\n(Flooded vs Non-Flooded Scenario)')
plt.xlabel('Client ID')
plt.ylabel('Number of Messages')
plt.grid(True)
plt.legend()
plt.xticks()

# Adjust layout and save
plt.tight_layout()
plt.show()
```
